## Requirement 2: 
Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
# The logic is finding all boolean values of columns WITHOUT ‘Not assigned’ value' and use them to create a new dataframe.
df_pure_index = df['Borough'] != 'Not assigned'
df_pure_index
```

```python
# use the above collection of boolean value to create the new dataframe free of 'Not assigned' value.
df1 = df[df['Borough'] != 'Not assigned']
df1
```

```python
# Check if we have completely dropped all "Not assigned" values.
df1['Borough'] == 'Not assigned'
```

Requirement 2 met.

__

## Requirement 3: 
More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11  in the above table.

```python
# Combine the neighborhoods with same Postalcode
df2 = df1.groupby(by=['Postal Code', 'Borough'], sort=False).agg(','.join)
df2.reset_index(inplace=True)
df2
```

Requirement 3 met.

__

## Requirement 4: 
If a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough.

Example: M7A | Queen's Park | Not assigned

```python
# np.where(condition, new value, old value)
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned', df2['Borough'], df2['Neighbourhood']) 
df2
```

# Clustering

## 1. Import the csv file containing latitudes and longitudes of various neighbourhoods in Canada

```python
lat_lon = pd.read_csv('http://cocl.us/Geospatial_data')
lat_lon.head()
```

## 2. Merge the coordinate table to the original table.

```python
# pd.merge(to table, from table, on='link') ---Very SQLish syntax huh?
df3 = pd.merge(df2, lat_lon, on='Postal Code')
df3.head()
```

```python
print('At this stage, the dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df['Borough'].unique()),
        df.shape[0]
    )
)
```

## 3. Get all rows containing Toronto in their Borough.

```python
df4 =df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

## 4. Visualize the map using Folium.

```python
toronto_lat_lng = [43.651070,-79.347015] # This is the coordinate of Toronto.

map_toronto = folium.Map(location=[43.651070,-79.347015], zoom_start=10) # create a map of toronto
```

```python
# superimpose data on the map

for lat,lng,borough,neighbourhood in zip(df4['Latitude'], df4['Longitude'], df4['Borough'], df4['Neighbourhood']): # traverse all designated data in df4
    label = '{}, {}'.format(neighbourhood, borough) # create a dictionary with designated values
    label = folium.Popup(label, parse_html=True) # superimpose the data on the map
    folium.CircleMarker([lat,lng], # more detailed settings
                        radius=5, # size of the circle
                        popup=label, # location of the circle
                        color='blue', # color of the frame of the circle
                        fill=True, # fill the circle with color
                        fill_color='#3186cc', # color of the filling
                        fill_opacity=0.7, # transparency level of the filling
                        parse_html=False #  HTML parsing is basically: taking in HTML code and extracting relevant information like the title of the page, paragraphs in the page, headings in the page, links, bold text etc.
                       ).add_to(map_toronto) # superimpose all the settings above to the map
                       
map_toronto
```

## Note: If you can not see the map on Github, check out the README for the map.

## 5. Cluster neighbourhoods using [KMeans Clustering]

```python
k=5 # set K value, which decides how many clusters the data will be grouped into.
toronto_clustering = df4.drop(['Postal Code', 'Borough', 'Neighbourhood'], 1) # only use numerical data to perform clustering

kmeans = KMeans(n_clusters = k, random_state=0).fit(toronto_clustering) # perform Kmeans Clustering
```

```python
kmeans.labels_ # display the clustering result
```

```python
# add the result to df4
df4.insert(0,'Cluster Labels', kmeans.labels_) #DataFrameName.insert(loc, 'column_name', value, allow_duplicates = False[optional])
df4
```

```python
df4
```

## 6. Visualize the clustering

```python
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=10) # create a map of toronto

## set color scheme for the cluster

x = np.arange(k) 
# Return evenly spaced values within a given interval
#format: numpy.arange([start, ]stop, [step, ]dtype=None, *, like=None)

ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys))) 
# use matplotlib cm library to create different colors for different ys.
#np.linspace returns evenly spaced numbers over a specified interval.

rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
## add markers to the map
markers_colors = [] # create empty list for markers colors

# superimpose data on the map

for lat,lng,neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']): # traverse all designated data in df4
    label = ' Cluster ' + str(cluster) # name the label. Example: Cluster 1, Cluster 2...
    label = folium.Popup(label, parse_html=True) # superimpose the cluster on the map
    folium.CircleMarker([lat,lng], # more detailed settings
                        radius=5, # size of the circle
                        popup=label, # location of the circle
                        color=rainbow[cluster-1], # color of the frame of the circle
                        fill=True, # fill the circle with color
                        fill_color=rainbow[cluster-1], # color of the filling
                        fill_opacity=0.7, # transparency level of the filling
                        parse_html=False #  HTML parsing is basically: taking in HTML code and extracting relevant information like the title of the page, paragraphs in the page, headings in the page, links, bold text etc.
                       ).add_to(map_clusters) # superimpose all the settings above to the map
                       
map_clusters
```
